Concatenate the three dataframes

```python
print("Shape df1", df1.shape)
print("Shape df2", df2.shape)
print("Shape df3", df3.shape)
```

```python
data = pd.concat([df1,df2,df3], axis=0)
data.shape
```

Perform the data cleaning operations mentioned so far in class

```python
data.drop(['education', 'number of open complaints'], axis = 1)
data.head(1)
```

#### Correct the values in the column `customer lifetime value`. They are given as a percent, so multiply (division) them by 100 and change `dtype` to `numerical` type.

```python
data.dtypes
```

```python
data['customer lifetime value'] = data['customer lifetime value'].str.replace('%', '')
```

```python
data['customer lifetime value'] = pd.to_numeric(data['customer lifetime value'], errors='coerce')
```

```python
data.dtypes['customer lifetime value']
```

```python
data['customer lifetime value'] = data['customer lifetime value']*100
data.head(10)
```
