**Sub Problem 1: How to aggregate the `amount_spent` for unique customers?**

```python
# your code here
total_spent = pd.DataFrame(orders.groupby(orders['CustomerID'])['amount_spent'].sum())
total_spent.head()
```

**Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?**

```python
percentile_95 = np.percentile(total_spent['amount_spent'], 95)
print(percentile_95)
percentile_75 = np.percentile(total_spent['amount_spent'], 75)
print(percentile_75)
```

**Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?**

```python
def customer(x):
    if x > percentile_95:
        return 'VIP'
    elif (x <= percentile_95) & (x >= percentile_75):
        return 'Preferred'
    else:
        return 'Unwanted'
    
total_spent['Customer Type'] = total_spent['amount_spent'].apply(customer)

total_spent
```

## Q2: How to identify which country has the most VIP Customers?

```python
#if we want details only on the returned orders
data = pd.merge(left = orders,
                right = total_spent,
                how = 'inner', 
                left_on = "CustomerID", 
                right_on= "CustomerID")
display(data.shape)
data.head()
```

```python
data['Customer Type'].value_counts()
```

```python
vip_countries = data[data['Customer Type'] == 'VIP']
vip_countries
```

```python
vip_countries['Country'].value_counts()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
type_c = ['VIP', 'Preferred']
vip_pref_countries = data[data['Customer Type'].isin(type_c) & data['Country']]
vip_pref_countries
```
