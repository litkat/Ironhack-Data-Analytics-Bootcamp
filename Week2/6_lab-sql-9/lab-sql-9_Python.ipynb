#### Create a Python connection with SQL database

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'  
engine = create_engine(connection_string)
```

#### 1. How many distinct (different) actors' last names are there?

```python
data1 = pd.read_sql_query('SELECT COUNT(DISTINCT last_name) AS "number of different actors last names"\
FROM sakila.actor', engine)
data1.head()
```

#### 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
data2 = pd.read_sql_query('SELECT *, \
CASE \
WHEN DATE_FORMAT(rental_date, "%W") IN ("Saturday", "Sunday") then "weekend" \
WHEN DATE_FORMAT(rental_date, "%W") IN ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday") then "workday" \
ELSE "INCORRECT DATE" \
END AS "day_type" \
FROM sakila.rental', engine)
data2.sample(10)
```

#### 3. Get all films with ARMAGEDDON in the title.

```python
data3 = pd.read_sql_query('SELECT * FROM sakila.film WHERE title LIKE "%ARMAGEDDON%"', engine)
data3.head()
```

#### 4. Get 10 the longest films.

```python
data4 = pd.read_sql_query('SELECT * FROM sakila.film ORDER BY length DESC LIMIT 10', engine)
data4.head()
```

#### 5. How many films include Behind the Scenes content?

```python
data5 = pd.read_sql_query('SELECT count(film_id) FROM sakila.film WHERE special_features LIKE "%Behind the Scenes%"', engine)
data5.head()
```

#### 6. Which kind of movies (rating) have a mean duration of more than two hours?

```python
data6 = pd.read_sql_query('SELECT rating, ROUND(AVG(length),2) \
FROM sakila.film \
GROUP BY rating \
HAVING ROUND(AVG(length),2) > 120', engine)
data6.head()
```

#### 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
data7 = pd.read_sql_query('SELECT title, length, DENSE_RANK() OVER (ORDER BY length DESC) "rank" \
FROM sakila.film \
WHERE length IS NOT null AND length > 0 \
ORDER BY length DESC', engine)
data7.head()
```
